Visualization of health data set

```python
# Author: Sushant Raj Kushawaha
# Roll No.: ACE080BCT089

# Date:2024/12/22
# Description: This is the visualization of health data set using matplotlib library of python.


import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)
df = load_dataset("./dataset.csv")
df.head(6)
```

1. Line Plot: Age vs Maximum Heart Rate (Thalach)

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='age', y='thalach', marker='o')
plt.title("Age vs Maximum Heart Rate (Thalach)")
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate (Thalach)")
plt.show()
```

--> Findings: The maximum heart rate achieved (thalach) generally decreases with age, showing an inverse relationship.

2. Bar Graph: Average Cholesterol Level by Sex


```python
plt.figure(figsize=(10, 6))
df.groupby('sex')['chol'].mean().plot(kind='bar', color=['blue', 'red'], alpha=0.7)
plt.title("Average Cholesterol Level by Sex")
plt.xlabel("Sex (0=Female, 1=Male)")
plt.ylabel("Average Cholesterol Level")
plt.xticks(rotation=0)
plt.show()
```

--> Findings: Males have slightly higher average cholesterol levels compared to females in this dataset.

3. Pie Chart: Distribution of Target (Heart Disease Presence)

```python
plt.figure(figsize=(8, 8))
df['target'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'salmon'], startangle=90)
plt.title("Distribution of Heart Disease Presence (Target)")
plt.ylabel("")
plt.show()
```

--> Findings: A higher proportion of individuals in the dataset have heart disease (target=1) compared to those without it (target=0).

4. Scatter Plot: Cholesterol vs Resting Blood Pressure

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='chol', y='trestbps', hue='target', alpha=0.7)
plt.title("Cholesterol vs Resting Blood Pressure")
plt.xlabel("Cholesterol")
plt.ylabel("Resting Blood Pressure")
plt.legend(title="Heart Disease")
plt.show()
```

--> Findings: There is no strong correlation between cholesterol and resting blood pressure, but individuals with heart disease (target=1) tend to cluster around lower resting blood pressures.


5. Histogram: Distribution of Age

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='purple', alpha=0.7)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

--> Findings: The age distribution is concentrated around the mid-50s, indicating most individuals in the dataset are middle-aged.


6. Heatmap: Correlation Matrix

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

--> Findings: The strongest correlations are between related cardiovascular metrics, such as resting blood pressure and oldpeak, or age and cholesterol.
